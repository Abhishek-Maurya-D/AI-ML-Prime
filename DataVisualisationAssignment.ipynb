```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
# 2 - line plot

sns.relplot(
    data = penguins,
    x = "flipper_length_mm",
    y = "body_mass_g",
    kind = "line",
    hue = "species"
)
```

```python
# 3 - box plot

tips = sns.load_dataset("tips")
sns.boxplot(
    data = tips,
    x = "day",
    y = "total_bill",
    hue = "sex"
)
tips
```

```python
# 4 - scatter plot

sns.scatterplot(
    data = tips,
    x = "total_bill",
    y = "tip",
    hue = "sex",
    style = "sex",
    size = "size"
)
plt.title("Total Bill Amount VS Tips")
plt.xlabel("Total Bill")
plt.ylabel("Tip Amount")
```

```python
tips
```

```python
# 5 - heatmap

tips_pivot = tips.pivot_table(
    index="day", 
    columns="time", 
    values="tip",
    aggfunc="mean"
)
tips_pivot
sns.heatmap(
    tips_pivot,
    annot=True
)
```
