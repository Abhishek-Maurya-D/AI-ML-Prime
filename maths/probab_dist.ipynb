```python
# Binomial Distribution

from scipy.stats import binom

n = 5
k = 3
p = 0.5

prob = binom.pmf(k,n,p)
print(prob)   # 5/16
```

```python
# Uniform Distribution

import numpy as np
import matplotlib.pyplot as plt

# values from 0 to 10
values = np.random.uniform(0, 10, 1_00_000)

# plotting a histograms - because more uniform with more values
plt.hist(values, bins=200, density=True, alpha=0.5)
plt.title("Continuous Uniform Distribution between 0 and 10")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
# Normal Distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 70 # mean
sigma = 10 # standard deviation

# x-axis values
x = np.linspace(30, 110, 1000)

# Normal PDF
y = norm.pdf(x, mu, sigma)

# Plot the bell curve
plt.plot(x, y, color="black", linewidth=2, label="Normal Distribution")

# Shade 68% region (  )
x1 = np.linspace(mu-sigma, mu+sigma, 1000)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color='green', alpha=0.7, label="68% region (1)")

# Shade 95% region (  2)
x2 = np.linspace(mu-2*sigma, mu+2*sigma, 1000)
plt.fill_between(x2, norm.pdf(x2, mu, sigma), color='yellow', alpha=0.6, label="95% region (2)")

# Shade 99.7% region (  3)
x3 = np.linspace(mu-3*sigma, mu+3*sigma, 1000)
plt.fill_between(x3, norm.pdf(x3, mu, sigma), color='red', alpha=0.5, label="99.7% region (3)")

# Labels and title
plt.title("Normal Distribution of Students Scores")
plt.xlabel("Score")
plt.ylabel("Probability Desnsity")
plt.legend()
plt.show()
```
