```python
# Grouping & Aggregation function on data
# groupby
# mean, sum, max, min, count, etc
# agg()

df.groupby("country")["income"].mean()
df.groupby("country")["income"].max()
df.groupby("country")["income"].min()

df.groupby("gender")["income"].mean()
df.groupby("gender")["income"].max()

df.groupby("country")["income"].agg(["mean", "max", "min"])
df.groupby("country")["income"].aggregate(["mean", "max", "min"])
df.groupby("country")["income"].agg(Average_Salary = "mean", Minimum_Salary = "min", Maximum_Salary = "max")

df.groupby("country").agg({
    "income" : "mean",
    "age" : "mean"
})

df.groupby("country").agg(
    Average_Income = ("income", "mean"),
    Average_Age = ("age", "mean")
)
```

```python
# Melt and Pivot
df = pd.DataFrame({
    "country" : ["USA", "USA", "India", "India"],
    "year" : [2020, 2021, 2020, 2021],
    "sales" : [100, 120, 90, 110],
    "profit" : [20, 25, 18, 22],
    "tax" : [10,20, 30, 40]
})
df
```

```python
melted_df = df.melt(
    id_vars = ["country", "year"],
    value_vars = ["sales", "profit", "tax"],
    var_name = "metrics",
    value_name = "value"
)
melted_df
```

```python
#pivot operations

original_df = melted_df.pivot(
    index = ["country", "year"],
    columns = "metrics",
    values = "value"
)
original_df
```

```python
# Merging & Joining Data

df_customers = pd.DataFrame({
    "customer_id" : [1,2,3,4],
    "name" : ["Abhshiek", "Aman", "Ankit", "Anuj"]
})

df_orders = pd.DataFrame({
    "order_id" : [101, 102, 103, 104],
    "customer_id" : [2, 1, 4, 5],
    "amount" : [250, 120, 300, 100]
})

pd.merge(df_customers, df_orders, on="customer_id") # common data - inner join
pd.merge(df_customers, df_orders, on="customer_id", how="left") # common data - left join
pd.merge(df_customers, df_orders, on="customer_id", how="right") # common data - right join
pd.merge(df_customers, df_orders, on="customer_id", how="outer") # common data - outer join
```

```python
# Concatentation

df1 = pd.DataFrame({
    "id" : [1,2,3],
    "name" : ["Adam", "Eve", "Bob"]
})

df2 = pd.DataFrame({
    "id" : [4,5,6],
    "name" : ["Charlie", "Fester", "Gilian"]
})

pd.concat([df2, df1]) # row wise 
pd.concat([df2, df1], ignore_index=True) # row wise 
pd.concat([df2, df1], axis = 1) # col wise 
```
